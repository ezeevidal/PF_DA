```python
# # Funci√≥n para calcular el porcentaje de nulos por columna
def calcular_porcentaje_nulos(df):
    porcentaje_nulos = (df.isnull().sum() / len(df))
    return porcentaje_nulos

# # Aplicar la funci√≥n a cada DataFrame y mostrar resultados
print("Porcentaje de datos nulos en cada archivo:\n")
print("2017 Purchase Prices Dec:")
print(calcular_porcentaje_nulos(df_2017_prices), "\n")

print("Big Inventory Final:")
print(calcular_porcentaje_nulos(df_big_inventory), "\n")
print("End Inventory Final:")
print(calcular_porcentaje_nulos(df_end_inventory), "\n")

print("Invoice Purchases:")
print(calcular_porcentaje_nulos(df_invoice_purchases), "\n")

print("Purchases Final:")
print(calcular_porcentaje_nulos(df_purchases_final), "\n")

print("Sales Final:")
print(calcular_porcentaje_nulos(df_sales_final), "\n")
```

```python
import pandas as pd
from sqlalchemy import create_engine

# Leer los CSVs
df_2017_prices = pd.read_csv('2017PurchasePricesDec.csv')
df_big_inventory = pd.read_csv('BigInvFinal.csv')
df_end_inventory = pd.read_csv('EndInvFinal.csv')
df_invoice_purchases = pd.read_csv('InvoicePurchases.csv')
df_purchases_final = pd.read_csv('PurchasesFinal.csv')
df_sales_final = pd.read_csv('SalesFinal.csv')

# Crear la conexi√≥n a la base de datos (usando SQLite en este caso)
engine = create_engine('sqlite:///mydatabase.db')  # Cambia esto por tu base de datos, si usas MySQL o PostgreSQL

# Exportar los DataFrames a tablas SQL
df_2017_prices.to_sql('purchase_prices_2017', con=engine, if_exists='replace', index=False)
df_big_inventory.to_sql('big_inventory', con=engine, if_exists='replace', index=False)
df_end_inventory.to_sql('end_inventory', con=engine, if_exists='replace', index=False)
df_invoice_purchases.to_sql('invoice_purchases', con=engine, if_exists='replace', index=False)
df_purchases_final.to_sql('purchases_final', con=engine, if_exists='replace', index=False)
df_sales_final.to_sql('sales_final', con=engine, if_exists='replace', index=False)

print("Datos exportados a la base de datos SQL exitosamente.")
```

```python
import pandas as pd

# Lista de tablas y sus rutas actualizadas
files = {
    "df_big_inventory": "big_inventory_modified.csv",
    "df_end_inventory": "end_inventory_modified.csv",
    "df_invoice_purchases": "invoice_purchases_modified.csv",
    "df_purchase_prices_2017": "purchase_prices_2017_modified.csv",
    "df_purchases_final": "purchases_final_modified.csv",
    "df_sales_final": "sales_final_modified.csv"
}

# Diccionario para almacenar tipos de datos
data_types = {}

# Cargar y analizar cada tabla
for name, path in files.items():
    try:
        df = pd.read_csv(path)
        data_types[name] = {
            "columns": list(df.columns),
            "dtypes": dict(df.dtypes),
            "shape": df.shape
        }
    except FileNotFoundError:
        data_types[name] = "Archivo no encontrado."
    except Exception as e:
        data_types[name] = f"Error: {str(e)}"

# Mostrar los resultados
for table, details in data_types.items():
    print(f"Tabla: {table}")
    if isinstance(details, dict):
        print(f"- Columnas: {details['columns']}")
        print(f"- Tipos de datos: {details['dtypes']}")
        print(f"- Dimensiones: {details['shape']}")
    else:
        print(f"- {details}")
    print()
```

```python
import pyodbc #Intalar pyodbc (pip install pyodbc)
import pandas as pd

# Funci√≥n para conectarse a la base de datos
def connect_to_sql_server(server, database, username, password):
    try:
        conn = pyodbc.connect(
            f"DRIVER={{ODBC Driver 17 for SQL Server}};"
            f"SERVER={server};"
            f"DATABASE={database};"
            f"UID={username};"
            f"PWD={password}"
        )
        print("Conexi√≥n exitosa a la base de datos.")
        return conn
    except Exception as e:
        print(f"Error al conectarse a la base de datos: {e}")
        return None

# Funci√≥n para actualizar datos
def update_database(conn, query, params=None):
    try:
        cursor = conn.cursor()
        if params:
            cursor.execute(query, params)
        else:
            cursor.execute(query)
        conn.commit()
        print("Base de datos actualizada correctamente.")
    except Exception as e:
        print(f"Error al actualizar la base de datos: {e}")

if __name__ == "__main__":
    # Detalles de conexi√≥n
    server = "servidor"
    database = "base de datos"
    username = "usuario"
    password = "contrase√±a"

    # Conexi√≥n
    conn = connect_to_sql_server(server, database, username, password)

    if conn:
        # Consulta para actualizar datos
        query = """
        UPDATE NombreDeLaTabla
        SET Columna = ?
        WHERE OtraColumna = ?
        """
        
        # Par√°metros para la consulta
        params = ("NuevoValor", "Condici√≥n")
        
        # Llamada a la funci√≥n para actualizar
        update_database(conn, query, params)
        
        # Cierre de la conexi√≥n
        conn.close()
        print("Conexi√≥n cerrada.")
```
